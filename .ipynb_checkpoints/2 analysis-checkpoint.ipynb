```python
import pandas as pd
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import hddm
import kabuki
```

```python
dfs=pd.read_csv("/home/jovyan/yikang1020/Ostwald_OSF/dfs_raw")
```

```python
def count_outvalues(df):
    iqr=df.quantile(0.75)-df.quantile(0.25)
    u_value=df.quantile(0.75)+1.5*iqr
    l_value=df.quantile(0.25)-1.5*iqr
    count=0
    for i in df:
        if (i>u_value)|(i<l_value):
            count += 1
    return count
def trans_outvalues(df):
    iqr=df.quantile(0.75)-df.quantile(0.25)
    u_value=df.quantile(0.75)+1.5*iqr
    l_value=df.quantile(0.25)-1.5*iqr
    for i,value in enumerate(df):
        if value>=u_value:
            df[i]=u_value
        elif value<=l_value:
            value[i]=l_value
    return df
```

```python
dfs['response_time']=trans_outvalues(dfs['response_time'])
plt.hist(dfs['response_time'],bins=20)
plt.show()
plt.boxplot(dfs.response_time)
plt.show()
dfs.describe()
```

```python
model=ols('response_time~C(response_corr)',data=dfs).fit()
anova_table=anova_lm(model,type=2)
pd.DataFrame(anova_table)
print(pairwise_tukeyhsd(dfs['response_time'],dfs['response_corr']))
```

```python
x=[1,0]
err=y=[dfs.loc[dfs['response_corr']==1,'response_time'].sem(),
dfs.loc[dfs['response_corr']==0,'response_time'].sem()]
y=[dfs.loc[dfs['response_corr']==1,'response_time'].mean(),
dfs.loc[dfs['response_corr']==0,'response_time'].mean()]
plt.bar(x,y,yerr=err,width=0.2)
```

```python
import sys
import time
from IPython.display import clear_output

def wait_watching_stdout(ar, dt=15):
    """
    ar: vmap output of the models being run 
    dt: number of seconds between checking output, you can make is shorter or longer.
    """

    while not ar.ready():
        stdouts = ar.stdout
        if not any(stdouts):
            continue
        clear_output()
        print('-' * 30)
        print("%.3fs elapsed" % ar.elapsed)
        print("")
        for out in ar.stdout: print(out);
        sys.stdout.flush()
        time.sleep(dt)
```

```python
def run_m(id):
    print('running model (depends on stim) %i'%id);
    
    import hddm
    
    exp_name = 'bias'
    model_tag = 'm'
    

    dbname = '/home/jovyan/yikang1020/Ostwald_OSF/dfs_' + exp_name + '_' + model_tag + '_chain_%i.db'%id 
   
    mname  = '/home/jovyan/yikang1020/Ostwald_OSF/dfs_' + exp_name + '_' + model_tag + '_chain_%i'%id    
    fname  = '/home/jovyan/yikang1020/Ostwald_OSF/dfs_1.csv'
    data = hddm.load_csv(fname)
    
    m = hddm.HDDM(data,depends_on={'v': 'prioritization_cue'})
    m.find_starting_values()
    m.sample(2000, burn=500,dbname=dbname, db='pickle') 
    m.save(mname)
    
    return m
```

```python
from ipyparallel import Client

v = Client()[:]

start_time = time.time()  

jobs = v.map(run_m, range(4)) 
wait_watching_stdout(jobs)

m_stim_list = jobs.get()

print("\nRunning 4 chains used: %f seconds." % (time.time() - start_time))
```

```python
m_stim_all = kabuki.utils.concat_models(m_stim_list) 
```

```python
v_WW, v_LL, v_WL = m_stim_all.nodes_db.node[['v(left)', 'v(right)', 'v(W)']]
hddm.analyze.plot_posterior_nodes([v_WW, v_LL, v_WL])
plt.xlabel('drift-rate')
plt.ylabel('Posterior probability')
plt.title('Posterior of drift-rate group means')
# plt.savefig('hddm_demo_fig_06.pdf')
```
